Ejercicio de clase que genere 10 notas de manera aleatorio y que calcula la media, la mejor y la peor nota, muestre cuantos estan aprobados, la mejor y peor nota y usando una funcion de analisis usando condicionales.

Sin usar pandas

```python
# Paso 1: Importar NumPy
import numpy as np

# Paso 2: Crear un array con las notas de 10 estudiantes generadas aleatoriamente (enteras)
notas = np.random.randint(0, 11, 10)  # Genera 10 notas enteras entre 0 y 10
print("Notas de los estudiantes:", notas)

# Paso 3: Definir la función de análisis
def analizar_notas(notas):
    # Calcular el promedio de las notas
    promedio = np.mean(notas)

    # Encontrar la mejor y peor nota
    mejor_nota = np.max(notas)
    peor_nota = np.min(notas)

    # Clasificar estudiantes en aprobados (>= 5) y reprobados (< 5)
    aprobados = notas >= 5  # Esto crea un array booleano
    num_aprobados = np.sum(aprobados)  # Suma los True (1) para contar aprobados
    num_reprobados = len(notas) - num_aprobados  # El resto son reprobados

    # Devolver los resultados en una tupla
    return promedio, mejor_nota, peor_nota, num_aprobados, num_reprobados

# Paso 4: Ejecutar la función y almacenar resultados
promedio, mejor_nota, peor_nota, num_aprobados, num_reprobados = analizar_notas(notas)

# Paso 5: Imprimir los resultados
print("\nResultados del análisis:")
print(f"Promedio de las notas: {promedio:.2f}")
print(f"Mejor nota: {mejor_nota}")
print(f"Peor nota: {peor_nota}")
print(f"Número de estudiantes aprobados: {num_aprobados}")
print(f"Número de estudiantes reprobados: {num_reprobados}")
```

Usando pandas

```python
# Paso 1: Importar Pandas y NumPy
import pandas as pd
import numpy as np

# Paso 2: Crear un DataFrame con las notas de 10 estudiantes generadas aleatoriamente (enteras)
notas = np.random.randint(0, 11, 10)  # Genera 10 notas enteras entre 0 y 10
df_notas = pd.DataFrame(notas, columns=["Notas"])
print("DataFrame de notas:")
print(df_notas)

# Paso 3: Definir la función de análisis
def analizar_notas(df):
    # Calcular el promedio de las notas
    promedio = df["Notas"].mean()

    # Encontrar la mejor y peor nota
    mejor_nota = df["Notas"].max()
    peor_nota = df["Notas"].min()

    # Clasificar estudiantes en aprobados (>= 5) y reprobados (< 5)
    aprobados = df[df["Notas"] >= 5]  # Filtra los aprobados
    num_aprobados = len(aprobados)  # Cuenta el número de aprobados
    num_reprobados = len(df) - num_aprobados  # El resto son reprobados

    # Devolver los resultados en una tupla
    return promedio, mejor_nota, peor_nota, num_aprobados, num_reprobados

# Paso 4: Ejecutar la función y almacenar resultados
promedio, mejor_nota, peor_nota, num_aprobados, num_reprobados = analizar_notas(df_notas)

# Paso 5: Imprimir los resultados
print("\nResultados del análisis:")
print(f"Promedio de las notas: {promedio:.2f}")
print(f"Mejor nota: {mejor_nota}")
print(f"Peor nota: {peor_nota}")
print(f"Número de estudiantes aprobados: {num_aprobados}")
print(f"Número de estudiantes reprobados: {num_reprobados}")
```
