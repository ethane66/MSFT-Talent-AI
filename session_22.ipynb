```python
# Parte 1: Preparación del entorno
# Ejecutar en Google Colab
!pip install diffusers transformers accelerate safetensors

# Parte 2: Cargar modelo de difusión preentrenado (Stable Diffusion)
from diffusers import StableDiffusionPipeline
import torch

pipe = StableDiffusionPipeline.from_pretrained(
    "CompVis/stable-diffusion-v1-4",
    torch_dtype=torch.float16,
    revision="fp16",
    use_auth_token=False  # Cambiar a True si se tiene un token de Hugging Face
).to("cuda")

# Parte 3: Generar una imagen a partir de texto (prompt)
prompt = "A futuristic city floating in the clouds, digital art, 4k"
image = pipe(prompt).images[0]
image.show()

# Se recomienda cambiar el prompt y probar diferentes variaciones
```
