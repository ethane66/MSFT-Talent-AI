```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Configuración de estilo para los gráficos
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)
plt.rcParams['font.size'] = 12

# Cargar los datos
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
titanic_df = pd.read_csv(url)

# Limpieza de datos
titanic_df = titanic_df.drop(columns=['Cabin', 'PassengerId', 'Ticket', 'Name'])
titanic_df['Age'] = titanic_df['Age'].fillna(titanic_df['Age'].median())
titanic_df['Embarked'] = titanic_df['Embarked'].fillna(titanic_df['Embarked'].mode()[0])

# Crear una columna para identificar niños
titanic_df['IsChild'] = titanic_df['Age'] < 18

# 1. Supervivencia por Género
plt.figure(figsize=(8, 6))
sns.countplot(x='Sex', hue='Survived', data=titanic_df, palette='pastel')
plt.title('Supervivencia por Género', fontsize=16)
plt.xlabel('Género', fontsize=14)
plt.ylabel('Número de Pasajeros', fontsize=14)
plt.legend(title='Sobrevivió', loc='upper right', labels=['No', 'Sí'])
plt.show()

# 2. Supervivencia por Clase
plt.figure(figsize=(8, 6))
sns.countplot(x='Pclass', hue='Survived', data=titanic_df, palette='pastel')
plt.title('Supervivencia por Clase', fontsize=16)
plt.xlabel('Clase', fontsize=14)
plt.ylabel('Número de Pasajeros', fontsize=14)
plt.legend(title='Sobrevivió', loc='upper right', labels=['No', 'Sí'])
plt.xticks(ticks=[0, 1, 2], labels=['Primera', 'Segunda', 'Tercera'])
plt.show()

# 3. Supervivencia por Edad
plt.figure(figsize=(10, 6))
sns.histplot(x='Age', hue='Survived', data=titanic_df, kde=True, palette='pastel', multiple='stack')
plt.title('Supervivencia por Edad', fontsize=16)
plt.xlabel('Edad', fontsize=14)
plt.ylabel('Número de Pasajeros', fontsize=14)
plt.legend(title='Sobrevivió', labels=['No', 'Sí'])
plt.show()

# 4. Supervivencia por Género y Clase
g = sns.catplot(x='Pclass', hue='Survived', col='Sex', kind='count', data=titanic_df, palette='pastel', height=5, aspect=1.2)
g.set_titles('{col_name}')
g.fig.suptitle('Supervivencia por Género y Clase', y=1.05, fontsize=16)
g.set_axis_labels('Clase', 'Número de Pasajeros')
g.set_xticklabels(['Primera', 'Segunda', 'Tercera'])
g._legend.set_title('Sobrevivió')
for ax in g.axes.flat:
    for p in ax.patches:
        ax.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='bottom')
plt.show()

# 5. Supervivencia de Niños
plt.figure(figsize=(8, 6))
sns.countplot(x='IsChild', hue='Survived', data=titanic_df, palette='pastel')
plt.title('Supervivencia de Niños', fontsize=16)
plt.xlabel('¿Es Niño?', fontsize=14)
plt.ylabel('Número de Pasajeros', fontsize=14)
plt.legend(title='Sobrevivió', loc='upper right', labels=['No', 'Sí'])
plt.xticks(ticks=[0, 1], labels=['Adulto', 'Niño'])
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Configuración de estilo para los gráficos
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)
plt.rcParams['font.size'] = 12

# Cargar los datos de propinas
tips_df = sns.load_dataset('tips')

# Mostrar las primeras filas del dataset
print(tips_df.head())

# 1. ¿Qué día se dan las propinas más altas?
plt.figure(figsize=(8, 6))
sns.barplot(x='day', y='tip', data=tips_df, palette='pastel', estimator=sum)
plt.title('Propinas por Día de la Semana', fontsize=16)
plt.xlabel('Día de la Semana', fontsize=14)
plt.ylabel('Total de Propinas', fontsize=14)
plt.show()

# 2. ¿Hay diferencias entre hombres y mujeres en la cantidad de propina que dejan?
plt.figure(figsize=(8, 6))
sns.boxplot(x='sex', y='tip', data=tips_df, palette='pastel')
plt.title('Distribución de Propinas por Género', fontsize=16)
plt.xlabel('Género', fontsize=14)
plt.ylabel('Propina', fontsize=14)
plt.show()

# 3. ¿Cómo se relaciona la cantidad de la cuenta con la propina?
plt.figure(figsize=(8, 6))
sns.scatterplot(x='total_bill', y='tip', data=tips_df, hue='sex', palette='pastel')
plt.title('Relación entre Total de la Cuenta y Propina', fontsize=16)
plt.xlabel('Total de la Cuenta', fontsize=14)
plt.ylabel('Propina', fontsize=14)
plt.legend(title='Género')
plt.show()

# 4. ¿El tamaño del grupo afecta la cantidad de propina?
plt.figure(figsize=(8, 6))
sns.boxplot(x='size', y='tip', data=tips_df, palette='pastel')
plt.title('Distribución de Propinas por Tamaño del Grupo', fontsize=16)
plt.xlabel('Tamaño del Grupo', fontsize=14)
plt.ylabel('Propina', fontsize=14)
plt.show()
```

Narración Mejorada con los Nuevos Gráficos
1. Supervivencia por Género
El gráfico muestra que las mujeres tuvieron una tasa de supervivencia mucho más alta que los hombres. Esto refleja la política de "mujeres y niños primero" que se implementó durante la evacuación. Mientras que la mayoría de las mujeres sobrevivieron, la mayoría de los hombres no lo lograron.

2. Supervivencia por Clase
La clase social fue un factor determinante. Los pasajeros de primera clase tuvieron una tasa de supervivencia significativamente mayor que los de segunda y tercera clase. Esto sugiere que el acceso a los botes salvavidas y la ubicación en el barco jugaron un papel crucial.

3. Supervivencia por Edad
El histograma muestra que los niños (menores de 18 años) tuvieron una mayor probabilidad de sobrevivir, especialmente los de primera y segunda clase. Sin embargo, los bebés y niños muy pequeños enfrentaron mayores dificultades, lo que se refleja en una ligera disminución en la tasa de supervivencia en ese grupo.

4. Supervivencia por Género y Clase
Este gráfico combina género y clase, revelando una doble desigualdad. Las mujeres de primera clase tuvieron la tasa de supervivencia más alta, mientras que los hombres de tercera clase tuvieron la más baja. Esto subraya cómo la combinación de género y clase socioeconómica influyó en las probabilidades de supervivencia.

5. Supervivencia de Niños
Los niños, en general, tuvieron una tasa de supervivencia más alta que los adultos. Sin embargo, los niños de tercera clase enfrentaron mayores obstáculos, lo que sugiere que la clase social también influyó en la supervivencia de los más pequeños.
