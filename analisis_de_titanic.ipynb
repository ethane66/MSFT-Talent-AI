```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Configuracin de estilo para los grficos
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)
plt.rcParams['font.size'] = 12

# Cargar los datos
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
titanic_df = pd.read_csv(url)

# Limpieza de datos
titanic_df = titanic_df.drop(columns=['Cabin', 'PassengerId', 'Ticket', 'Name'])
titanic_df['Age'] = titanic_df['Age'].fillna(titanic_df['Age'].median())
titanic_df['Embarked'] = titanic_df['Embarked'].fillna(titanic_df['Embarked'].mode()[0])

# Crear una columna para identificar nios
titanic_df['IsChild'] = titanic_df['Age'] < 18

# 1. Supervivencia por Gnero
plt.figure(figsize=(8, 6))
sns.countplot(x='Sex', hue='Survived', data=titanic_df, palette='pastel')
plt.title('Supervivencia por Gnero', fontsize=16)
plt.xlabel('Gnero', fontsize=14)
plt.ylabel('Nmero de Pasajeros', fontsize=14)
plt.legend(title='Sobrevivi', loc='upper right', labels=['No', 'S'])
plt.show()

# 2. Supervivencia por Clase
plt.figure(figsize=(8, 6))
sns.countplot(x='Pclass', hue='Survived', data=titanic_df, palette='pastel')
plt.title('Supervivencia por Clase', fontsize=16)
plt.xlabel('Clase', fontsize=14)
plt.ylabel('Nmero de Pasajeros', fontsize=14)
plt.legend(title='Sobrevivi', loc='upper right', labels=['No', 'S'])
plt.xticks(ticks=[0, 1, 2], labels=['Primera', 'Segunda', 'Tercera'])
plt.show()

# 3. Supervivencia por Edad
plt.figure(figsize=(10, 6))
sns.histplot(x='Age', hue='Survived', data=titanic_df, kde=True, palette='pastel', multiple='stack')
plt.title('Supervivencia por Edad', fontsize=16)
plt.xlabel('Edad', fontsize=14)
plt.ylabel('Nmero de Pasajeros', fontsize=14)
plt.legend(title='Sobrevivi', labels=['No', 'S'])
plt.show()

# 4. Supervivencia por Gnero y Clase
g = sns.catplot(x='Pclass', hue='Survived', col='Sex', kind='count', data=titanic_df, palette='pastel', height=5, aspect=1.2)
g.set_titles('{col_name}')
g.fig.suptitle('Supervivencia por Gnero y Clase', y=1.05, fontsize=16)
g.set_axis_labels('Clase', 'Nmero de Pasajeros')
g.set_xticklabels(['Primera', 'Segunda', 'Tercera'])
g._legend.set_title('Sobrevivi')
for ax in g.axes.flat:
    for p in ax.patches:
        ax.annotate(f'{int(p.get_height())}', (p.get_x() + p.get_width() / 2., p.get_height()), ha='center', va='bottom')
plt.show()

# 5. Supervivencia de Nios
plt.figure(figsize=(8, 6))
sns.countplot(x='IsChild', hue='Survived', data=titanic_df, palette='pastel')
plt.title('Supervivencia de Nios', fontsize=16)
plt.xlabel('Es Nio?', fontsize=14)
plt.ylabel('Nmero de Pasajeros', fontsize=14)
plt.legend(title='Sobrevivi', loc='upper right', labels=['No', 'S'])
plt.xticks(ticks=[0, 1], labels=['Adulto', 'Nio'])
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Configuracin de estilo para los grficos
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)
plt.rcParams['font.size'] = 12

# Cargar los datos de propinas
tips_df = sns.load_dataset('tips')

# Mostrar las primeras filas del dataset
print(tips_df.head())

# 1. Qu da se dan las propinas ms altas?
plt.figure(figsize=(8, 6))
sns.barplot(x='day', y='tip', data=tips_df, palette='pastel', estimator=sum)
plt.title('Propinas por Da de la Semana', fontsize=16)
plt.xlabel('Da de la Semana', fontsize=14)
plt.ylabel('Total de Propinas', fontsize=14)
plt.show()

# 2. Hay diferencias entre hombres y mujeres en la cantidad de propina que dejan?
plt.figure(figsize=(8, 6))
sns.boxplot(x='sex', y='tip', data=tips_df, palette='pastel')
plt.title('Distribucin de Propinas por Gnero', fontsize=16)
plt.xlabel('Gnero', fontsize=14)
plt.ylabel('Propina', fontsize=14)
plt.show()

# 3. Cmo se relaciona la cantidad de la cuenta con la propina?
plt.figure(figsize=(8, 6))
sns.scatterplot(x='total_bill', y='tip', data=tips_df, hue='sex', palette='pastel')
plt.title('Relacin entre Total de la Cuenta y Propina', fontsize=16)
plt.xlabel('Total de la Cuenta', fontsize=14)
plt.ylabel('Propina', fontsize=14)
plt.legend(title='Gnero')
plt.show()

# 4. El tamao del grupo afecta la cantidad de propina?
plt.figure(figsize=(8, 6))
sns.boxplot(x='size', y='tip', data=tips_df, palette='pastel')
plt.title('Distribucin de Propinas por Tamao del Grupo', fontsize=16)
plt.xlabel('Tamao del Grupo', fontsize=14)
plt.ylabel('Propina', fontsize=14)
plt.show()
```

Narracin Mejorada con los Nuevos Grficos
1. Supervivencia por Gnero
El grfico muestra que las mujeres tuvieron una tasa de supervivencia mucho ms alta que los hombres. Esto refleja la poltica de "mujeres y nios primero" que se implement durante la evacuacin. Mientras que la mayora de las mujeres sobrevivieron, la mayora de los hombres no lo lograron.

2. Supervivencia por Clase
La clase social fue un factor determinante. Los pasajeros de primera clase tuvieron una tasa de supervivencia significativamente mayor que los de segunda y tercera clase. Esto sugiere que el acceso a los botes salvavidas y la ubicacin en el barco jugaron un papel crucial.

3. Supervivencia por Edad
El histograma muestra que los nios (menores de 18 aos) tuvieron una mayor probabilidad de sobrevivir, especialmente los de primera y segunda clase. Sin embargo, los bebs y nios muy pequeos enfrentaron mayores dificultades, lo que se refleja en una ligera disminucin en la tasa de supervivencia en ese grupo.

4. Supervivencia por Gnero y Clase
Este grfico combina gnero y clase, revelando una doble desigualdad. Las mujeres de primera clase tuvieron la tasa de supervivencia ms alta, mientras que los hombres de tercera clase tuvieron la ms baja. Esto subraya cmo la combinacin de gnero y clase socioeconmica influy en las probabilidades de supervivencia.

5. Supervivencia de Nios
Los nios, en general, tuvieron una tasa de supervivencia ms alta que los adultos. Sin embargo, los nios de tercera clase enfrentaron mayores obstculos, lo que sugiere que la clase social tambin influy en la supervivencia de los ms pequeos.
