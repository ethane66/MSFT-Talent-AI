Generar el prompt dinámico.
Pasarlo a GPT-2 como entrada y generar el informe con estos parámetros:
Decodificar y mostrar el informe generado:


```python
# -------------------------------------------
# Generador de Informes Ejecutivos con GPT-2
# -------------------------------------------


# Paso 1: Instalar Hugging Face Transformers (si no está instalado)
!pip install transformers --quiet


# Paso 2: Importar las librerías necesarias
from transformers import GPT2LMHeadModel, GPT2Tokenizer
import torch


# Paso 3: Cargar modelo y tokenizador GPT-2
tokenizer = GPT2Tokenizer.from_pretrained("gpt2")
model = GPT2LMHeadModel.from_pretrained("gpt2")
model.eval()  # Modo evaluación (no entrenamiento)


# Paso 4: Definir los datos del cliente (puedes cambiar estos valores)
empresa = "NovaTech"
sector = "tecnología educativa"
evento = "ha incrementado su número de usuarios en un 47% durante el último trimestre"
objetivo = "expandirse a América Latina durante 2025"


# Paso 5: Crear el prompt dinámico usando f-strings
prompt = (
    f"La empresa {empresa}, del sector {sector}, ha experimentado recientemente un "
    f"resultado significativo: {evento}. De cara al futuro, su objetivo principal es "
    f"{objetivo}. A continuación, se presenta un informe ejecutivo con recomendaciones:\n\n"
)


# Paso 6: Codificar el prompt en tokens que entienda GPT-2
inputs = tokenizer.encode(prompt, return_tensors="pt")


# Paso 7: Generar texto usando el modelo GPT-2
outputs = model.generate(
    inputs,
    max_length=250,        # Longitud total del texto (incluye prompt)
    do_sample=True,        # Activar aleatoriedad
    top_k=50,              # Solo considerar las 50 mejores opciones
    top_p=0.95,            # Nucleus sampling
    temperature=0.7        # Controla la creatividad (1.0 = neutral)
)


# Paso 8: Decodificar los tokens generados a texto legible
informe = tokenizer.decode(outputs[0], skip_special_tokens=True)


# Paso 9: Mostrar el informe generado
print("="*80)
print("📄 INFORME GENERADO POR GPT-2:\n")
print(informe)
print("="*80)


```
